```python
# Import modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')

#plt.rcParams['figure.figsize'] = (20, 10)
#plt.rcParams['font.family'] = "serif"
```

```python
df = pd.read_csv('D:\\Study\\DataScience\\Data\\MachineLearningData\\HighwayData\\vartest.csv')
```

```python
df['Yr'] = pd.to_datetime(df['Yr'], format = '%Y')
df = df.set_index('Yr')
df.index = pd.DatetimeIndex(df.index)
```

```python
df.head()
```

```python
#creating the train and validation set
train = df[:int(0.8*(len(df)))]
valid = df[int(0.8*(len(df))):]
```

```python
valid
```

```python
#fit the model
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
model = VAR(endog=train)
model_fit = model.fit()
```

```python
# make prediction on validation
prediction = model_fit.forecast(model_fit.y, steps=len(valid))
```

```python
#converting predictions to dataframe
cols = df.columns
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[cols])
```

```python
#converting predictions to dataframe
pred = pd.DataFrame(index=range(0,len(prediction)),columns=[cols])
for j in range(0,len(cols)):
    for i in range(0, len(prediction)):
        pred.iloc[i][j] = prediction[i][j]
```

```python
#check rmse
from sklearn.metrics import mean_squared_error

for i in range(0,len(cols)):
    #print(cols[i])
    mse = mean_squared_error(pred.iloc[:,i], valid.iloc[:,i])
    rmse = np.sqrt(mse)
    print('RMSE for', cols[i],'is:',format(np.round(rmse,3)))
```

```python
#make final predictions
model = VAR(endog=df)
model_fit = model.fit()
yhat = model_fit.forecast(model_fit.y, steps=1)
print(yhat)
```
