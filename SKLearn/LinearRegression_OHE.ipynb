```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("D:/Study/Data _Science/LinearRegression/LinearRegressionOHE.csv")

df.head()
```

```python
df['State'].unique()
```

```python
#x = df.loc[:, df.columns != 'Profit']

#y = df.loc[:, df.columns == 'Profit']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
dfle = df.copy()
```

```python
dfle.State = le.fit_transform(dfle.State)

dfle
```

```python
dfle
```

```python
x = dfle.loc[:, dfle.columns != 'Profit'].values

y = dfle.loc[:, dfle.columns == 'Profit'].values

x
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(categorical_features= [3])
```

```python
x = ohe.fit_transform(x).toarray()
```

```python
x = x[:, 1:]

x
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size = 0.3, random_state = 0)
```

```python
linearRegressor = LinearRegression()
```

```python
linearRegressor.fit(xTrain,yTrain)
```

```python
yPredict = linearRegressor.predict(xTest)
```

```python
from sklearn.metrics import r2_score

print(r2_score(yTest,yPredict))
```
