```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# ignore warnings

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# load cancer dataset

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
```

```python
# split datasets into training and testing datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size = 0.3, random_state = 0)
```

```python
# Fit the training data using Logistic Regression model

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)

# print the training and testing score:

print('Training Score: {:0.3f}'.format(lr.score(X_train, y_train)))
print('Testing Score: {:0.3f}'.format(lr.score(X_test, y_test)))
```

```python
# Lets use value of C = 100

lr100 = LogisticRegression(C = 100)

lr100.fit(X_train, y_train)

# print the training and testing scores with C = 100

print('Training Score with C = 100: {:0.3f}'.format(lr100.score(X_train, y_train)))
print('Testing Score with C = 100: {:0.3f}'.format(lr100.score(X_test, y_test)))
```

```python
# Lets use value of C = 0.01

lr01 = LogisticRegression(C = 0.01)

lr01.fit(X_train, y_train)

# print the training and testing scores with C = 0.01

print('Training Score with C = 0.01: {:0.3f}'.format(lr01.score(X_train, y_train)))
print('Testing Score with C = 0.01: {:0.3f}'.format(lr01.score(X_test, y_test)))
```

```python
plt.plot(lr.coef_.T, 'o', label="C=1")
plt.plot(lr100.coef_.T, '^', label="C=100")
plt.plot(lr01.coef_.T, 'v', label="C=0.001")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.hlines(0, 0, cancer.data.shape[1])
plt.ylim(-5, 5)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.legend()
```
