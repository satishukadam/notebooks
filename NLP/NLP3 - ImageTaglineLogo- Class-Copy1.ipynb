```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read image

image = cv2.imread("D:\Study\DataScience\OpenCV\messi1.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# read logo
logo = cv2.imread("D:\Study\DataScience\OpenCV\logo.jpg")
logo = cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)
logo = cv2.resize(logo, (160, 80))
```

```python
class placeImageLogo:
    
    def __init__(self):
        self
    
    def placeLogo(self, image, logo, loc = 'bc'):
        
        assert image.shape[0] > logo.shape[0], "Height of main image is less than logo image"
        assert image.shape[1] > logo.shape[1], "Width of main image in less than logo image"
        assert loc in ['bc', 'tl', 'bl', 'tr', 'br'], "Please specify corrent location from 'bc', 'tl', 'bl', 'tr', 'br'"
        
        self.image = image
        self.logo = logo
        self.loc = loc
        
        self.x_offset = 50
        self.y_offset = 50
        
        if self.loc == 'br':
            image[image.shape[0]- logo.shape[0]-self.y_offset:image.shape[0]-self.y_offset, 
                  image.shape[1]- logo.shape[1]-self.x_offset:image.shape[1]-self.x_offset] = logo
            
        elif self.loc == 'tl':
            image[self.y_offset:self.y_offset+logo.shape[0], self.x_offset:self.x_offset+logo.shape[1]] = logo
            
        elif self.loc == 'bl':
            image[image.shape[0]- logo.shape[0]-self.y_offset:image.shape[0]-self.y_offset, 
                  self.x_offset:self.x_offset+logo.shape[1]] = logo
            
        elif self.loc == 'tr':
            image[self.y_offset:self.y_offset+logo.shape[0], 
                  image.shape[1]- logo.shape[1]-self.x_offset:image.shape[1]-self.x_offset] = logo
            
        elif self.loc == 'bc':
            image[image.shape[0]-logo.shape[0]-self.y_offset:image.shape[0]-self.y_offset,
      int(image.shape[1]-logo.shape[1])//2:int(image.shape[1]-logo.shape[1])//2 + logo.shape[1]] = logo
            
        
        return plt.imshow(image)
 
```
