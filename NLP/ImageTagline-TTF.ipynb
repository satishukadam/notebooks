```python
def split_tagline(tagline):
    words = tagline.split()
    split_length = len(words)//2 + 1
    list_of_words = [words[x:x+4] for x in range(0, len(words), split_length)]
    list_of_lines = [' '.join(list_of_words[i]) for i in range(0, len(list_of_words))]
    return list_of_lines
```
