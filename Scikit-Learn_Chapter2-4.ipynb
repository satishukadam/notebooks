<b> Build a Random Forest Classifier on make moons dataset </b>

```python
# import modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create dummy data from make moons

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples = 100, noise = 0.25, random_state = 2) 
```

```python
# split the data into training and testing datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Fit the Random Forest Classifier

rf = RandomForestClassifier(n_estimators=5, random_state = 1)

rf.fit(X_train, y_train)

# print training and testing scores

print('Training Score: {:0.3f}'.format(rf.score(X_train, y_train)))
print('Testing Score: {:0.3f}'.format(rf.score(X_test, y_test)))
```

<b> Build a Random Forest Classifier on breast cancer dataset </b>

```python
# load cancer data

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

X1_train, X1_test, y1_train, y1_test = train_test_split(cancer.data, cancer.target, test_size = 0.3, random_state = 0)
```

```python
# Fit the random forest classifier

rf1 = RandomForestClassifier(n_estimators = 100, random_state = 0)

rf1.fit(X1_train, y1_train)

# print training and testing scores

print('Training Score: {:0.3f}'.format(rf1.score(X1_train, y1_train)))
print('Testing Score: {:0.3f}'.format(rf1.score(X1_test, y1_test)))
```

```python
# Create a zipped list of feature names against its importance

feat_list = list(zip(cancer.feature_names, rf1.feature_importances_))
```

```python
import operator

# Create a sorted list 

feat_list.sort(key = operator.itemgetter(1))
```

```python
# print the plot of feature importance and feature names

plt.figure(figsize= (12,10))
plt.barh([x[0] for x in feat_list], [x[1] for x in feat_list] )
```

<b> Build a Gradient Boosting Classifier on breast cancer dataset </b>

```python
# import the module

from sklearn.ensemble import GradientBoostingClassifier

grb = GradientBoostingClassifier(random_state = 0)

# Fit the classifier

grb.fit(X1_train, y1_train)

# print the scores

print('Training Score: {:0.3f}'.format(grb.score(X1_train, y1_train)))
print('Testing Score: {:0.3f}'.format(grb.score(X1_test, y1_test)))
```

```python
# provide max depth parameter

grb1 = GradientBoostingClassifier(random_state = 0, max_depth=1)

# Fit the classifier

grb1.fit(X1_train, y1_train)

# print the scores

print('Training Score: {:0.3f}'.format(grb1.score(X1_train, y1_train)))
print('Testing Score: {:0.3f}'.format(grb1.score(X1_test, y1_test)))
```

```python
# provide learning rate parameter

grb2 = GradientBoostingClassifier(random_state = 0, learning_rate=0.01)

# Fit the classifier

grb2.fit(X1_train, y1_train)

# print the scores

print('Training Score: {:0.3f}'.format(grb2.score(X1_train, y1_train)))
print('Testing Score: {:0.3f}'.format(grb2.score(X1_test, y1_test)))
```

```python
# plot feature importance plot

feat_imp_grb = list(zip(cancer.feature_names, grb2.feature_importances_))

# sort the list as per feature importance

feat_imp_grb.sort(key= operator.itemgetter(1))

# plot
plt.figure(figsize=(10,8))
plt.barh([x[0] for x in feat_imp_grb], [x[1] for x in feat_imp_grb])
```
