```python
# Import basic modules

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('D:/Study/DataScience/RecommenderSystems/movie_testdataset.csv')
```

```python
df['features'] = df['genres'] + " " + df['keywords'] + " " + df['cast'] + " " + df['director']
```

```python
# Import modules required to create similarity matrix

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create count matrix
cv = CountVectorizer()

cv.fit_transform(df['features'])

cv_matrix = cv.fit_transform(df['features']).toarray()

# Create Similarity Matrix

similarity_matrix = cosine_similarity(cv_matrix)

print(similarity_matrix)
```

```python
# Function to return index of a movie if name of the movie is given

def get_index_from_title(name):
    try:
        return df[df['original_title'] == name ]['index'].values[0]
    except:
        print('Movie not found in list !!')
    
movie_index = get_index_from_title('Men in Black 3')

print(movie_index)
```

```python
# Once the index of the movie is known sort the corresponding index row in similarity matrix

similar_movies  = list(enumerate(similarity_matrix[movie_index]))

# sorted movie list with index number

def takeSecond(elem):
    return elem[1]

sorted_movie_list = sorted(similar_movies, key=takeSecond, reverse= True)
```

```python
# Now return top 5 movie names with Indexes

def get_title_from_index(index):
    return df[df['index'] == index ]['original_title'].values[0]

for movie in sorted_movie_list[0:5]:
    print(get_title_from_index(movie[0]))

    
```
