```python
# Import basic python libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Create sample data using sklearn

from sklearn.datasets.samples_generator import make_blobs

a, b = make_blobs(n_samples = 500, centers = 5, cluster_std = 0.65, random_state = 0)

plt.scatter(a[:, 0], a[:,1])

plt.show()
```

```python
# Import sklearn for KMeans

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5)

kmeans.fit(a)
```

```python
print(kmeans.cluster_centers_)

print(kmeans.inertia_)
```

```python
# Display the cluster centroids in different colours
labels = kmeans.labels_

li = labels.tolist()

lst
```

```python
df = pd.DataFrame(a)

df['label'] = li

df.head()
```

```python
def pltcolor(lstt):
    cols=[]
    for l in lstt:
        if l==0:
            cols.append('red')
        elif l== 1:
            cols.append('blue')
        elif l== 2:
            cols.append('yellow')
        elif l== 3:
            cols.append('orange')
        else:
            cols.append('green')
    return cols
# Create the colors list using the function above
cols=pltcolor(df.iloc[:,2])

plt.figure(figsize=(10,10))
plt.scatter(df.iloc[:,0],df.iloc[:, 1],s=40,c=cols) #Pass on the list created by the function here
plt.grid(True)


plt.show()
```
