```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('D:/Study/DataScience/OpenCV/carwithpeople1.jpg')

img_gr = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#plt.imshow(img1)

ret, thresh = cv2.threshold(img_gr, 127, 255, 0)

contours, heirarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
 
img_cn = cv2.drawContours(image, contours, -1, (255,0,0), 1)

plt.figure(figsize=(18,8))
plt.imshow(img_cn)
```

```python
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 

# fontScale 
fontScale = 2

# text to be placed on image
text = 'This is the tagline'
textsize = cv2.getTextSize(text, font, fontScale, 2)[0]

# origin 
org = (int((image.shape[1]-textsize[0])/2),int(image.shape[0]*15/100))
  
# Yellow font color in RGB 
color = (255,0,0) 
  
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.putText() method 
tagged_image = cv2.putText(image, text, org, font, fontScale, color, thickness, cv2.LINE_AA)
#boxed_image = cv2.rectangle(tagged_image, (org[0], org[1] - textsize[1]), (org[0] + textsize[0],org[1]), color, thickness)

# display tagged image
plt.figure(figsize=(10,12))
plt.imshow(tagged_image)

# save tagged image
# cv2.imwrite('/path/to/destination/image.png',image)

#tagged_image.shape
```

```python
for contour in contours:
    l, r, t, b, bb = [], [], [], [], []
    (x, y, w, h) = cv2.boundingRect(contour)
    leftmost = tuple(contour[contour[:,:,0].argmin()][0])
    rightmost = tuple(contour[contour[:,:,0].argmax()][0])
    topmost = tuple(contour[contour[:,:,1].argmin()][0])
    bottommost = tuple(contour[contour[:,:,1].argmax()][0])
    l.append(leftmost)
    r.append(rightmost)
    t.append(topmost)
    b.append(bottommost)
    #print(x,y, w ,h)
    #print(t)
    cv2.rectangle(tagged_image, (x,y), (x+w, y+h), color, thickness)
    if (x >= org[0] and x <= org[0] + textsize[0]) and (y >= org[1] - textsize[1] and y <= org[1]):
        print('Overlapping')
        print(x, y, w, h)
    else:
        print('Not Overlapping')
    
    
plt.figure(figsize=(10,12))    
plt.imshow(tagged_image)   
```
