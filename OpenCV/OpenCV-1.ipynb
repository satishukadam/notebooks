```python
import numpy as np 
import datetime
from datetime import date
   
# Create a VideoCapture object and read from input file 
cap = cv2.VideoCapture('D:/Study/DataScience/OpenCV/tree.avi') 
cap.set(cv2.CAP_PROP_FPS, 30)
fps = cap.get(cv2.CAP_PROP_FPS)
print(fps)
   
# Check if camera opened successfully 
if (cap.isOpened()== False):
    print("Error opening video  file") 
    
# Read until video is completed 
while(cap.isOpened()): 
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True: 
   
        # select font
        font = cv2.FONT_HERSHEY_SIMPLEX
        # get frame width and height
        width = str(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = str(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
        # define text to add to the frame
        #text = 'Width: ' + width + '\n' 'Height: ' + height
        w = 'Width: ' + width
        h = 'Height: ' + height
    
        # add text to the frame
        #frame = cv2.putText(frame, text, (10, 50), font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)
        frame = cv2.putText(frame, w, (10, 20), font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)
        frame = cv2.putText(frame, h, (10, 40), font, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
    
        # add date to the frame
        today = str(date.today())
        t = 'Day :' + today
        frame = cv2.putText(frame, t, (int(frame.shape[1]/2), 20), font, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        
        # Display the resulting frame 
        cv2.imshow('Frame', frame) 
   
        # Press Q on keyboard to  exit 
        if cv2.waitKey(2500) & 0xFF == ord('q'):
            break
   
  # Break the loop
    else:  
        break
        
# When everything done, release  
# the video capture object 
cap.release() 
   
# Closes all the frames 
cv2.destroyAllWindows()
```
